```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

df = pd.read_csv('customer churn.csv')
df.head()
                 
```

# Replacing blanks with 0 in total charges as data type also Changes Recorded.

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

### converting senior citizen value into yes or no to make understable

```python
def conv(value):
    if value==1:
      return "yes"
    else:
      return "no"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)
plt.title("Count of customer by  churn")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct = "%1.2f%%")
plt.title("Percentage Of Churned Costomer",fontsize = 10)
plt.show()
```

### For the given pie chart we can conclude that 26.54% of costomer have churn out.

```python
df_total = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Create the figure and axis
plt.figure(figsize=(4,4))
ax = df_total.plot(kind='bar', stacked=True, ax=plt.gca(), width=0.8, color=['#1f77b4', '#ff7f0e'])  # Customizing colors

# Annotate with percentages
for container in ax.containers:
    ax.bar_label(container, label_type='center', fmt='%.1f%%')  # Labels in the center of the bars
ax.set_xticklabels(['No', 'Yes'])
# Customizing labels and title
ax.set_xlabel('Senior Citizen')
ax.set_ylabel('Percentage')
ax.set_title('Senior Citizen Churn Distribution')

plt.show()
```

```python
df.columns.values
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set the number of rows and columns for the subplots
n_cols = 3  # Number of columns in the subplot grid
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten()  # Flatten the axes array for easy indexing

# Loop through the columns and create count plots
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(f'Count Plot of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

### Across most services, customers who did not churn outnumber those who did. Key patterns indicate that customers without additional features like security or backup tend to have higher churn rates, while those with fiber-optic Internet service also show notable churn.
